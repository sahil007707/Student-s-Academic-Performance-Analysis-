# Data Cleaning & Handling

```python
df.isna().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

# EDA & Visualization

```python
classifi_st=df.groupby("gender")["study_hours"].mean().reset_index()
classifi_st.plot(kind="bar",x="gender",y="study_hours",edgecolor="black")
plt.xlabel("Gender")
plt.title("Average Study Hours By Gender",size=13)
plt.tight_layout()
plt.show()
```

```python
classifi_sc=df.groupby("gender")["exam_score"].sum().reset_index()
classifi_sc.plot(kind="bar",x="gender",y="exam_score",edgecolor="black")
plt.xlabel("Gender")
plt.title("Exam Score By Gender",size=13)
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
# Plot: Study hours vs Exam score, sized by age, colored by gender
fig = px.scatter(
    df,
    x='study_hours',
    y='exam_score',
    color='gender',
    size='age',
    hover_data=['age'],
    title='Study Hours vs Exam Score by Age and Gender'
)

fig.update_layout(
    xaxis_title='Study Hours',
    yaxis_title='Exam Score',
    legend_title='Gender',
    title_x=0.5,
    template='plotly_white'
)

fig.show()
```

```python
cor=df.select_dtypes(include="number").corr()
plt.figure(figsize=(12,8))
sns.heatmap(data=cor,annot=True,fmt='.2f')
plt.title("Correlation Heatmap",size=13)
plt.show()
```
